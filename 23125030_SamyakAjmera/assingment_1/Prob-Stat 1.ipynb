Q1)a.

$$ Let \  np=\lambda= constant $$
 $$  p=\lambda/n  $$
   
 $$  \lim_{{n \to \infty}} P{(X=x)}= \binom{n}{x} * p^x*(1-p)^{n-x} $$
 $$   P(X=x)= \lim_{{n \to \infty}} \frac{n!}{x! (n-x)!} {(\lambda/n)}^x (1-\lambda/n)^{n-x}\\ = \lim_{{n \to \infty}} \frac{(n)(n-1)(n-2)...(n-x+1)}{x!} {(\lambda/n)}^x (1-\lambda/n)^{n-x} \\ $$
 
 $$ \\ \\ \lim_{{n \to \infty}}\frac{n!}{(n-x)!} * (\frac{1}{n})^x = (\frac{n}{n})\,\frac{(n-1)}{n}\,\frac{(n-2)}{n}....\frac{(n-x+1)}{n} \ \to 1  \\ \\\ $$
 
$$ \\ \lim_{{n \to \infty}}(1-\lambda/n)^{-x} \to 1 \\ $$

$$ \\ \lim_{{n \to \infty}}(1+ (-\lambda)/n)^{n} \to e^{-\lambda} \\ $$

$$ Thus \ from \ the \  above \ equations \ we \ get \ the \ Poisson \ distribution \ pmf \ as : \\ $$
   $$ P(X=x)= \frac {e^{-\lambda} \cdot\ (\lambda)^x}{x!} \\ $$

```python
from scipy.stats import poisson, binom
import numpy as np
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2)

mu = 4.6
x1 = np.arange(poisson.ppf(0.01, mu), poisson.ppf(0.99, mu))
ax1.plot(x1, poisson.pmf(x1, mu), 'bo', ms=5, label='Poisson PMF')
ax1.vlines(x1, 0, poisson.pmf(x1, mu), colors='b', lw=5, alpha=0.5)
ax1.legend(frameon=True)

n, p = 20, 0.23
x2 = np.arange(binom.ppf(0.01, n, p), binom.ppf(0.99, n, p))
ax2.plot(x2, binom.pmf(x2, n, p), 'ko', ms=5, label='Binomial PMF')
ax2.vlines(x2, 0, binom.pmf(x2,n,p), colors='k', lw=5, alpha=0.5)
ax2.legend(frameon=True)
ax2.legend()

plt.show()
```

```python
mean_p, var_p, skew_p, kurt_p = poisson.stats(mu, moments='mvsk')
print(mean_p, var_p, skew_p, kurt_p)

mean_b, var_b, skew_b, kurt_b = binom.stats(n,p, moments='mvsk')
print(mean_b, var_b, skew_b, kurt_b)
```

Thus from the above experiments it is clear that we can take binomial approximation to poisson only when n is large and p is small. For small n and large p there is significant difference specially in the variance, skewness and kurtosis.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

probabilities = np.array([1/8,3/16,1/16,1/8,5/16,0,3/16,0])

def generate_sample():
    digits = [2,3,1,2,5,0,3,0]
    sample = np.random.choice(digits, p=probabilities, size=10000)
    return sample

sample= [generate_sample() for i in range(10000)]
sample_means = [np.mean(i) for i in sample]

sns.histplot(sample_means, kde=True)
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```

The Resultant graph corresponds to that of a normal distribution. This is independent of the orginal distribution. This can be proved using the central limit theorem (CLT).

```python
sample_mean = np.mean(sample)

sample_variance = np.var(sample, ddof=1)  # Using ddof=1 for unbiased estimate

print("Mean of the experimental distribution:", sample_mean)
print("Variance of the experimental distribution:", sample_variance)
```

The mean and variance of the original data and that calculated using the CLT are quit similar to that calculated by creationg 10000 samples and the above histogram. CLT is proved.
